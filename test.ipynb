```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

```python
# 1. 加载 .env 文件中的环境变量
# 这行代码会去当前目录寻找 .env 文件，并读取里面的配置
load_dotenv()

# 2. 从环境变量中读取你的配置
api_key = os.getenv("QWEN_API_KEY")
base_url = os.getenv("QWEN_BASE_URL")
model_name = os.getenv("QWEN_CHAT_MODEL") # 例如 qwen-plus

# 3. 检查API Key是否成功加载
if not api_key:
    print("❌ 错误：未找到 QWEN_API_KEY，请检查 .env 文件。")
    exit()
else:
    print("✅ 已找到 API Key，准备发起请求...")

# 4. 初始化 OpenAI 客户端 (因为阿里云提供了兼容的接口)
try:
    client = OpenAI(
        api_key=api_key,
        base_url=base_url,
    )
    print("✅ 客户端初始化成功。")
except Exception as e:
    print(f"❌ 客户端初始化失败: {e}")
    exit()

# 5. 发起一个简单的聊天请求进行测试
print(f"🔄 正在向模型 {model_name} 发送测试请求...")
try:
    completion = client.chat.completions.create(
        model=model_name,
        messages=[
            {"role": "user", "content": "你好，请用一句话介绍一下你自己。"}
        ],
        timeout=30  # 设置30秒超时，避免长时间等待
    )

    # 6. 打印回复内容
    if completion.choices:
        reply = completion.choices[0].message.content
        print("\n" + "="*50)
        print("✅ 测试成功！模型回复如下：")
        print("="*50)
        print(f"模型: {reply}")
        print("="*50)
    else:
        print("❌ 请求成功但未获取到回复内容。")

except Exception as e:
    print(f"\n❌ 请求失败，发生错误: {e}")
    print("\n可能的原因：")
    print("1. API Key 不正确或已过期")
    print("2. 网络连接问题，无法访问 aliyun.com")
    print("3. 账户余额不足或免费额度已用完")
    print("4. .env 文件中的模型名称配置错误")
```
